```python
import numpy as np
import matplotlib.pyplot as plt
from fitter import Fitter

# Simulación de datos de lluvia (mm)
# Muchos valores pequeños y pocos grandes
np.random.seed(42)
lluvia = np.random.gamma(shape=1.8, scale=4.2, size=50)

# Histograma
plt.hist(lluvia, bins=10)
plt.title("Histograma de precipitación diaria (mm)")
plt.xlabel("mm de lluvia")
plt.ylabel("Frecuencia")
plt.show()

# Ajuste de distribuciones
f = Fitter(lluvia, distributions=['gamma', 'expon', 'norm'])
f.fit()

# Mostrar mejor ajuste
print(f.get_best(method='sumsquare_error'))

# Graficar ajuste
f.summary()
```
